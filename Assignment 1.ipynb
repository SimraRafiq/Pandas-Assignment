```python
import pandas as pd
from glob import glob
```

```python
us_census=sorted(glob("Downloads/data/states*.csv"))
us_census
```

```python
us_census=pd.concat((pd.read_csv(file).assign(filename=file)
          for file in us_census),ignore_index=False)
```

```python
us_census
```

```python
us_census.isnull().sum()
```

```python
us_census=us_census.fillna(method="ffill")
us_census.head()
```

```python
us_census.isnull().sum()
```

```python
us_census.columns
```

```python
us_census.dtypes
```

```python
us_census.head()
```

```python
Gender_pop=us_census["GenderPop"].str.split("_",n=1,expand=True)
Gender_pop
```

```python
Gender_pop["the_men_column"]=Gender_pop[0].str.replace("M","")
Gender_pop["the_men_column"]=pd.to_numeric(Gender_pop["the_men_column"])
Gender_pop["the_men_column"]
```

```python
Gender_pop["the_women_column"]=Gender_pop[1].str.replace("F","")
Gender_pop["the_women_column"]=pd.to_numeric(Gender_pop["the_women_column"])
Gender_pop["the_women_column"]
```

```python
Gender_pop["the_women_column"].dtypes
```

```python
Gender_pop["the_women_column"].isnull()
```

```python
Gender_pop["the_women_column"]=Gender_pop["the_women_column"].fillna(method="ffill")
Gender_pop["the_women_column"]
```

```python
the_income_column=us_census["Income"].str.split("$",n=1,expand=True)
the_income_column
```

```python
the_income_column["the_income"]=the_income_column[1].str.replace("$","")
the_income_column["the_income"]=pd.to_numeric(the_income_column["the_income"])
the_income_column["the_income"]
```

```python
the_income_column["the_income"].dtypes
```

```python
the_income_column["the_income"].isnull()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(Gender_pop["the_women_column"],the_income_column["the_income"],color="r")
plt.show() 
```
